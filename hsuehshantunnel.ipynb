```python
data = pd.read_csv('hhtt.csv')
data
```

```python
#for i in range(0, data.shape[0]):
#    if data.vsrid[i] == 3:
#        print(i)
```

```python
data = data.loc[data.loc[:, 'carid'] == 'S', :]
#data
data_x = data.iloc[:, [5, 7, 8]]
```

```python
kmeans_fit = cluster.KMeans(n_clusters = 6).fit(data_x)
```

```python
cluster_labels = kmeans_fit.labels_
print(cluster_labels)
```

```python
data_y = pd.DataFrame(cluster_labels)
data_y
```

```python
data_x_train = data_x.iloc[range(0, int(data_x.shape[0] * 0.7)), :]
data_x_test  = data_x.iloc[range(int(data_x.shape[0] * 0.7), int(data_x.shape[0])), :]
data_y_train = data_y.iloc[range(0, int(data_y.shape[0] * 0.7)), :]
data_y_test  = data_y.iloc[range(int(data_y.shape[0] * 0.7), int(data_y.shape[0])), :]
```

```python
svc = svm.SVC()
svc_fit = svc.fit(data_x_train , data_y_train)
```

```python
test_y_predicted = svc.predict(data_x_test)

# 績效
accuracy = metrics.accuracy_score(data_y_test , test_y_predicted)
print(accuracy)
```
